#**World Map Chart**
This world map reveals the global spread of violent discipline. Countries with higher rates are vividly highlighted, making it easy to recognize regional disparities. The visualization showcases how child violence is not limited to one area but is a widespread issue requiring international attention and immediate policy action.

```python
import pandas as pd
import geopandas as gpd
from plotnine import *

# Load your UNICEF indicator data
df_indicator = pd.read_csv('/content/unicef_indicator_2 (1).csv')

# Rename columns
df_indicator.rename(columns={'country': 'Country', 'obs_value': 'Most_recent_value'}, inplace=True)

# Load World Map shapefile
world = gpd.read_file("world_data/ne_110m_admin_0_countries.shp")

# Merge datasets
world_merged = world.merge(df_indicator, how='left', left_on='NAME', right_on='Country')

# Set figure size explicitly
plot = (
    ggplot(world_merged.dropna(subset=['Most_recent_value'])) +
    geom_map(aes(fill='Most_recent_value'), color='black') +
    scale_fill_gradient(low="lightblue", high="red") +
    labs(title='World Map: % of Children Experiencing Violent Discipline',
         fill='% Violent Discipline') +
    theme_minimal() +
    theme(figure_size=(10, 3))  # Proper plot size adjustment
)

# Display the plot
plot
```

#**Bar Chart**
The bar chart highlights the top 15 countries with the highest rates of violent discipline. Ranking these countries visually brings into focus where the crisis is most urgent. It challenges us to look beyond averages and prioritize targeted interventions in nations where children face the greatest risk.

```python
#| label: bar-chart
top_countries = df_indicator.nlargest(15, 'Most_recent_value')

(
    ggplot(top_countries, aes(x='reorder(Country, -Most_recent_value)', y='Most_recent_value')) +
    geom_bar(stat='identity', fill='red') +
    labs(x='Country', y='% Violent Discipline', title='Top 15 Countries by Violent Discipline') +
    theme_minimal() +
    theme(axis_text_x=element_text(rotation=45, hjust=1), figure_size=(12, 7))
)
```

#**Scatterplot with Linear Regression**
The scatterplot compares country performance, plotting the relationship between countries and violence rates. Although socio-economic patterns can influence child welfare, the data shows a scattered reality. The regression line offers a trend direction, but variations suggest cultural and political factors significantly impact child discipline practices globally.

```python
#| label: scatter-time
import warnings
warnings.filterwarnings('ignore')

(
    ggplot(df_indicator, aes(x='time_period', y='Most_recent_value')) +
    geom_point(color='red', size=2, alpha=0.7) +
    geom_smooth(method='lm', color='blue', se=False) +
    labs(title='Trend of Violent Discipline over Time',
         x='Year',
         y='% Violent Discipline') +
    theme_minimal() +
    theme(figure_size=(10, 6))
)

```

#**Density Plot**
The density plot visualizes the spread of violent discipline rates across countries. Most nations cluster around moderate violence levels, but significant peaks suggest worrying concentrations. This visualization reveals that while extreme cases are fewer, the majority of countries face substantial child welfare challenges requiring urgent global intervention and systemic change.

```python
#| label: density-plot
(
    ggplot(df_indicator, aes(x='Most_recent_value')) +
    geom_density(fill='red', color='darkblue') +
    labs(title='Density of Violent Discipline Rates Across Countries',
         x='% Violent Discipline', y='Density') +
    theme_minimal() +
    theme(figure_size=(8, 6))
)
```
